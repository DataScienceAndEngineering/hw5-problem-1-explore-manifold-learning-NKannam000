```python
from keras.datasets import cifar10
from sklearn.decomposition import PCA
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
from sklearn.manifold import Isomap
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.manifold import MDS
from sklearn.manifold import SpectralEmbedding
from sklearn.manifold import TSNE
import umap.umap_ as umap
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

### Cleaning up and reshaping data

```python
reshape_Xtrain = X_train.reshape(50000,3072)
reshape_Xtest = X_test.reshape(10000,3072)
reshape_Ytrain = y_train.reshape(50000,1)
reshape_Ytest = y_test.reshape(10000,1)
```

```python
reshape_Xtrain.shape
```

```python
reshape_Ytrain.shape
```

```python
print('Training Images: {}'.format(X_train.shape))
print('Testing Images: {}'.format(X_test.shape))
print('Training Label: {}'.format(y_train.shape))
print('Testing Label: {}'.format(y_test.shape))
```

### List of the labels

```python
labels = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship','truck']
```

### PCA

```python
pca = PCA(n_components = 2)
X_train_pca = pca.fit_transform(reshape_Xtrain[:5000])
```

```python
figure(figsize=(8, 6), dpi=80)
plt.rcParams["image.cmap"] = "turbo"
plot = plt.scatter(X_train_pca[:,0], X_train_pca[:,1], c=reshape_Ytrain[:5000], s = 15)
plt.legend(handles=plot.legend_elements()[0],labels=labels)
plt.show()
```

### Isomap

```python
embedding = Isomap(n_components=2)
X_train_I = embedding.fit_transform(reshape_Xtrain[:5000])
```

```python
figure(figsize=(8, 6), dpi=80)
plot = plt.scatter(X_train_I[:,0], X_train_I[:,1], c=reshape_Ytrain[:5000], s=15)
plt.legend(handles=plot.legend_elements()[0], labels=labels)
plt.show()
```

### Locally Linear Embedding

```python
embedding_ll = LocallyLinearEmbedding(n_components=2)
X_train_ll = embedding_ll.fit_transform(reshape_Xtrain[:5000])
```

```python
figure(figsize=(8, 6), dpi=80)
plot = plt.scatter(X_train_ll[:,0], X_train_ll[:,1], c=reshape_Ytrain[:5000], s=15)
plt.legend(handles=plot.legend_elements()[0], labels=labels)
plt.show()
```

### Multidimensional scaling

```python
embedding_MDS = MDS(n_components=2, normalized_stress='auto')
X_train_MDS = embedding_MDS.fit_transform(reshape_Xtrain[:5000])
```

```python
figure(figsize=(8, 6), dpi=80)
plot = plt.scatter(X_train_MDS[:,0], X_train_MDS[:,1], c=reshape_Ytrain[:5000], s=15)
plt.legend(handles=plot.legend_elements()[0], labels=labels)
plt.show()
```

### Spectral Embedding

```python
embedding_se = SpectralEmbedding(n_components=2)
X_train_se = embedding_se.fit_transform(reshape_Xtrain[:5000])
```

```python
figure(figsize=(8, 6), dpi=80)
plot = plt.scatter(X_train_se[:,0], X_train_se[:,1], c=reshape_Ytrain[:5000], s=15)
plt.legend(handles=plot.legend_elements()[0], labels=labels)
plt.show()
```

### T-distributed Stochastic Neighbor Embedding

```python
embedding_tsne = TSNE(n_components=2)
X_train_tnse = embedding_tsne.fit_transform(reshape_Xtrain[:5000])
```

```python
figure(figsize=(8, 6), dpi=80)
plot = plt.scatter(X_train_tnse[:,0], X_train_tnse[:,1], c=reshape_Ytrain[:5000], s=15)
plt.legend(handles=plot.legend_elements()[0], labels=labels)
plt.show()
```

### UMAP

```python
embedding_umap = umap.UMAP()
X_train_umap = embedding_umap.fit_transform(reshape_Xtrain[:5000])
```

```python
figure(figsize=(8, 6), dpi=80)
plot = plt.scatter(X_train_umap[:,0], X_train_umap[:,1], c=reshape_Ytrain[:5000], s=15)
plt.legend(handles=plot.legend_elements()[0], labels=labels)
plt.show()
```
